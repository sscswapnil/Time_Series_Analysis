```python
import pandas as pd
from fbprophet import Prophet
%matplotlib inline
pd.plotting.register_matplotlib_converters() # Required to plot the graph in my_case
```

```python
df = pd.read_csv(r'F:\DS\Time_Series_Analysis\[ FreeCourseWeb.com ] Udemy - Python for Time Series Data Analysis\01. Introduction\UDEMY_TSA_FINAL\Data\HospitalityEmployees.csv')
```

```python
df.columns = ['ds','y']
df['ds'] = pd.to_datetime(df['ds'])
```

```python
df.plot(x='ds',y='y',figsize=(12,8),label='Employees')
```

```python
train = df.iloc[:336]
test = df.iloc[336:]
```

```python
m = Prophet()
m.fit(train)
future = m.make_future_dataframe(periods=12,freq='MS')
forecast = m.predict(future)
```

```python
from fbprophet.plot import add_changepoints_to_plot
fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), m, forecast)  # gca= get_current_access
```
